```python
import pandas as pd

boston_df = pd.read_csv('./datasets/Boston.csv')
boston_df
```

```python
x = boston_df.drop(columns=['Unnamed: 0','medv'], axis=1)
y = boston_df['medv']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=00)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

x_train = mms.fit_transform(x_train)
x_test = mms.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('Variance score : {0:.3f}'.format(r2_score(y_test,y_pred)))
```

```python
print('ì ˆíŽ¸ ê°’: ', lr.intercept_)
print('íšŒê·€ ê³„ìˆ˜ê°’:', np.round(lr.coef_,1))
```
