```python
import pandas as pd
```

```python
bike_df = pd.read_csv('./datasets/bikerental.csv')
```

```python
bike_df
```

```python
x = bike_df.drop(['datetime', 'casual', 'registered', 'count'], axis=1)
y = bike_df['count']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
bike_train_scalered = scaler.fit_transform(x_train)
bike_test_scalered = scaler.fit_transform(x_test)
```

```python
x_train = pd.DataFrame(bike_train_scalered, columns=x.columns)
x_train
```

```python
x_test = pd.DataFrame(bike_test_scalered, columns=x.columns)
x_test
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
# mean_squared_error = 오차의 제곱에 대한 평균값
from sklearn.metrics import mean_squared_error

# 학습
# rg.fit(bike_train_scalered, y_train)
# # 예측
# pred = rg.predict(bike_train_scalered, y_test)
# # 평가
# accuracy = accuracy_score(y_test, pred)
# accuracy

# rg.fit(bike_train_scalered, y_train)
# pred = rg.predict(bike_test_scalered)
# print('DecisionTreeClassifier 정확도 : {0:.4f}'.format(accuracy_score(y_test, pred)))

param = {'max_depth':[6,8,10,12],
        'n_estimators':[100],
        'min_samples_leaf':[8,12,18],
        'min_samples_split':[8,16,20]}

rf_clf = RandomForestRegressor(random_state=156,n_jobs=-1)
grid_cv = GridSearchCV(rf_clf,param_grid=param,cv=2,scoring='neg_mean_squared_error',n_jobs=-1)
grid_cv.fit(x_train,y_train)
pred = grid_cv.predict(x_test)
best_dt = grid_cv.best_estimator_

mean_squared_error = mean_squared_error(y_test, pred)
mean_squared_error
```

```python
best_dt
```

```python
# from sklearn.svm import SVR
# clf = svm.SVR(gamma=0.001, C=100.)
# clf.fit(x_train, y_train)
# pred = clf.predict(x_test)
# from sklearn.metrics import accuracy_score
# print(accuracy_score(y_test, pred))

from sklearn.svm import SVR
svr = SVR()
pram = {'C':[1,10],
       'epsilon':[0.1,0.2,0.3]}
grid = GridSearchCV(svr,param_grid=pram,cv=2,scoring="neg_mean_squared_error")
grid.fit(x_train,y_train)
best_regressor = grid.best_estimator_
```

```python
best_regressor
```

```python
from sklearn.metrics import mean_squared_error

results_dt = best_dt.predict(x_test)
results_svr = best_regressor.predict(x_test)

print(f'랜덤포레스트 mse : {mean_squared_error(y_test, results_dt, squared=False)}')
print(f'SVM mse : {mean_squared_error(y_test, results_svr, squared=False)}')
```
